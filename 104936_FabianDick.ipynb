##  1.1 An Application of Deep Reinforcement Learning to Algorithmic Trading
### By Thubazt Théate, Damine Ernst in Expert Systems with Applications Volume 173, 1 July 2021
SJR Rating: https://www.scimagojr.com/journalsearch.php?q=24201&tip=sid&clean=0

### 1.1.1 Topic and purpose of the paper

Since algorithmic trading is already well known in 2021 Théate et al. are facing the problem of how to design a algorithm based on artificial intelligence techniques in order to compete with popular algorithmic trading strategies which use a set of mathematical rules, such as trend following or mean reversion strategies.
For this purpose the authors use the Deep Q-Network algorithm. The framework should not only outperform algorithmic trading or quantitative trading, which is used as a synonym, but it should also solve the problem of what, when, how, at which price and at which quantity a trade should be performed.

The reasons why this paper was picked for a write up are on one hand the interesting modifications of the agent and it´s broad section about the performance assessment, e. g. the algorithm is tested within different sectors.

#### 1.1.2 Technical and methodology 

Since the environment the agent is interacting with is very complex, the authors equip the agent with several features to reduce the complexity of financial markets:



*  $S_t$ which represents the state information of the agent at time step t, which are the current trading position, number of owned shares and available cash
* $D_t$ is the information contained within the Open-High-Low-Close-Volume (OHLCV) 
* $T_t$ is the information regarding the time step t, which is date, weekday, and time
* $I_t$ includes several multiple technical indicators regarding the stock market targeted at time step t, such as the moving average convergence divergence (MACD), the relative strength index (RSI) or the average directional index (ADX).
*  $M_t$ gathers macroeconomic information such as interest or exchange rates.
*  $N_t$ covers data obtained by news from e. g. social media or newspapers.
*  $E_t$ contains extra useful information such as rumours or experts´advice.

Even if the set of variables is quite big, the authors point out that the information available is extremely restricted compared to the environments complexity.


##### 1.1.2.1 Modifications

To be able outperforming the benchmark models the authors Q-learning framework includes several modifications and improvements and distinguishes the algorithm from a classical Deep Q-learning network (DQN). A sample of the most relevant adjustments is given hereafter:

###### The architecture of the deep neural network (DNN)

> The architecture of the neural network uses a classical feedforward deep neural network (DNN)  with rectified linear unit activation functions, the so called Leaky ReLU, which is an addition to classical ReLUs and solves the problem of information leaks. While ReLUs are generally used to handle nonlinearity, the Leaky ReLU allows a small gradient when the unit is not active.


###### Double DQN

> To solve the overestimations from traditional DQN algorithm the authors modify the framework with a so called double DQN presented by van Hasselt et. al in 2015. This modification is based on the decomposition of the target max operation into action selection and action evaluation. This is done by evaluating the policy according to the network and use a target network to estimate its value. The result is a combination of policy based and value based algorithms , such as the actor-to-critic (A2C) approach.

###### Huber loss

> The Huber loss H is an alternative to the classical mean squared error (MSE) loss and stabilizes the training phase. The Huber loss overgoes the possibility of large errors triggered by the MSE, since the MSE penalizes large errors significantly which leads to a significant change in the target. The Huber loss therefore updates the DNN slower and more stable and is a good tradeoff between the MSE and the mean absolute error (MAE). The metric H gives the loss x the value $0.5x^2$ if $|x|≤1$ and $|x|-0.5$ otherwise to the normalized values. While the normalization of all inputs is done by a batch normalization layer.

###### Gradient clipping

> Another modification which relates to the model stability is the so-called gradient clipping technique. It controls the policy gradient update to ensure that the updated policy is not too different from the previous one. To fulfill this task the clip function clips the ratio $r_{t(θ)}$ to be within  $1 - ϵ$ and $1+ ϵ$, while the ratio $r_{t(θ)}$ is calculated by:

$$
r_{t(θ)} = \frac{\ \pi_{\boldsymbol{\theta}}(a_t | s_t)}{\pi_{\boldsymbol{\theta}_{old}}(a_t | s_t)}
$$

Then $r_{t(θ)}$ is multipled with the Q value to calculate the clipped policy objective. Then the objective function takes the minimum of the clipped and the normal objective.

*   Buy and hold (B&H)

While the first two approaches can be described as passive strategies, TF and MR are active strategies and issue multiple changes in trading positions (long/short) over the time periods.

TF thereby is concerned with the identification and the follow-up of significant market trends while MR is based on going back to the previous average price despite a upcoming or ongoing trend. In other words, a TF strategy makes profit when a MR strategy does not and vice versa. This is because MR denies and goes against the trend while TF follows it.

### 1.2 Deep learning for decision making and the optimization of socially responsible investments and portfolio
#### By Nhi N.Y.Vo, Xuezhong He, Shaowu Liu, Guandong Xu in Decision Support Systems Volume 124, September 2019
SJR Rating: https://www.scimagojr.com/journalsearch.php?q=21933&tip=sid&clean=0

The next write up replicates the intuition and the results of a paper which shows next to stock trading and preferred profitability of the agents action a more specific challenge. Also, it is chosen as it shows that reinforcement learning can be attached to traditional finance theories. 

#### 1.2.1 Topic and purpose of the paper

Vo et al. construct in this paper a framework which builds an optimized socially responsible investment portfolio which contains a Multivariate Bidirectional Long Short-Term Memory neural network. Next to traditional performance indicators the algorithm is also measured by the social impact of it´s portfolio built. The background of the paper is the increasing interest for financial investors and researchers to consider the environmental, social and governance issues of a company, e. g. refugees or global warming and the investment shift to companies which fulfill these requirements.
According to the authors there is a lack of quantitative approaches for investments with social responsibility while the current investors must read Corporate Social Responsibility Reports in order to detect possible companies, which is very time consuming.  To seek out companies with social and environmental impact on the one hand and on the other also make profitable investments is the aim of the algorithm proposed by Nhi N.Y. Voa, Xuezhong Heb, Shaowu Liua and Guandong Xua.

#### 1.2.2 Technical and methodology 
Since the authors try to combine an arithmetic problem, which is trading, with a topic that has mostly qualitative data they face the problem to bring these together. Therefore, they refer to the Global Reporting Initiatives (GRI) and the United Nation Sustainable Development Goals (SDGs) which provide standardized metrics and frameworks for companies. Such as Environmental, Social and Governance (ESG) metrics which can now be applied to address the portfolio building problem.
But even there is still enough data, the authors face certain problems. First the model must be able to predict multiple time series for stock returns. Second, existing portfolio optimization theories, such as the Mean-Variance approach do not fulfill the requirements of investments with social responsibility. Third, the portfolio must be able to adapt to market movements in order to be profitable but also achieve ESG rating objectives.
To solve these issue Vo et al. construct a manufactory approach captured in the the Deep Responsible Investment Portfolio (DRIP) model that consists of three main parts:


*  Multivariate Bi-Long-short-term-memory (BiLSTM) neural network to predict stock returns

*   MVP-ESG model for portfolio construction

*   Reinforcement learning approaches to retrain predictions and rebalance the portfolios 


##### 1.2.2.1 The Multivariate BiLSTM Neural Network (predicting stock returns)

For solving the challenge of predicting stock returns Vo et. al use a special type of Long Short-Term Memory (LSTM) networks. In general a LSTM network can learn long-term dependencies by using several so called gates. The input gate, which decides which values of the state are updated by the input signal, a forget gate, that decides which information of the previous state is remembered or forgotten and a output gate which allows the state to have or not an effect on other neurons. These three gates control the memory cell $C_t$ that preserves long-term information. The BiLSTM is a special kind of LSTM in which the data is presented to the gates once from the beginning to the end and once from the end to the beginning.  

##### 1.2.2.2 MVP-ESG model for portfolio construction

The traditional Mean Variance (MV) Portfolio has the components of maximizing return and minimizing risk. The raw version of this approach fulfills the requirement of profitable trades in combination with lower risk but lacks the ability of making Socially Responsible Investments (SRI). Therefore, Vo et. al adjust this approach based on the Sequential Least Squares Programming (SLSQP) method which is able to minimize a function of various variables and can be used within the context of trading for maximizing returns, minimizing volatility and maximizing ESG ratings. The output is the ESG-adjusted Sharpe Ratio S ̃, which is calculated by minimizing the negative Sharpe ratio with the portfolio ESG rating as a new variable within the objective function.


##### 1.2.2.3 DRIP Model with Reinforcement learning (retrain predictions and rebalance the portfolios)

In this part it is explained how reinforcement learning comes in place and how the ESG variable is implemented.

The BiLSTM neural networks and the MVP-ESG models explained above are combined into a single reinforcement learning model named DRIP (Deep Responsible Investment Portfolio). For this purpose a simple reward function is defined which is the sum of future discounted rewards:

$$
V^s_{\pi}= E(Re) = E[\sum_{t=0}γ^tRe_t | s_0 = s],
$$
while $s$ being the state and $γ^t$ the discount factor.

Further $Re$ is calculated by 

$$
Re = \sum_t=0γ^t S\frac{1}{MSE_t},
$$

while $S ̃ $ is the ESG-adjusted Sharpe ratio from the MVP-ESG model and MSE is the mean squares error of prediction.

After each time step $Δt$ the DRIP updates the prediction model with new stock prices and constructs a new portfolio with updated allocation weights regarding the portfolio performance and stock weights.

The authors used various ESG rating services for the adjustments which are unfortunately not all free available. However, Yahoo Finance offers ESG ratings which are publicly available, e. g. Visa Inc. (finance.yahoo.com/quote/V/sustainability/):

The used data are the daily closing prices of all Standard and Poor 500 (S&P500) stocks reaching from December 1988 to December 2018 with a data splitting ratio of 9:1, which means that the years from 1989 to 2015 is meant for training and the remaining years for testing. Further Vo et al. use a rolling forecast approach to split up the testing period into validation and test sets. The rolling is performed quarterly and, in a second, run yearly. 

Removing all companies which have a missing value in December 1988 leads to 262 companies in total. Another data preprocessing is due to the purpose of the DRIP, meaning that relevant investors do not invest in companies with low ESG ratings. #### 1.2.3 The empirical output and it´s evaluation

Since one of the main challenges the algorithm is facing is the prediction accuracy, the authors use four different measures. 

While using also classical metrices, the Mean Absolute Error (MAE) and the Root Mean Squared Error (RMSE), the authors also use the prediction accuracy of up- and downtrend forecasts (in %) and the Area Under the Curve (AUC) which is a metric obtained by the Receiver Operating Characteristic (ROC) curve. The ROC curve evaluates the specificities of the accuracy of the trend forecasts.

The predicted returns from the DRIP model were used as input for the MV-ESG model with the purpose of constructing social responsible investment portfolios. 

The portfolio returns where then compared to sustainable indexes, such as the Dow Jones Sustainability World Index (DJSI World) or the S&P500 ESG Factor Weighted Index (S&P500 ESG). Additionally the authors compared the RL approach with a Non-RL framework, in which the models were retrained after each testing period without learning from the previous period(s). The results show that RL significantly boosts the performance  regarding prediction accuracy, ESG score or financial measures such as the (adjuste) Sharpe ratio.


#### 1.2.4 Learnings and key take aways
The work presented shows that RL can construct socially responsible portfolios which outperform comparable indices. ## 1.3 Reinforcement learning applied to Forex trading
### by João Carapuço, Rui Neves and Nuno Horta

SJR Rating: https://www.scimagojr.com/journalsearch.php?q=18136&tip=sid&clean=0

The next write up replicates a paper which shows, next to the stock market, another application of reinforcement learning within the context of financial markets. The authors João Carapuço, Rui Neves and Nuno Horta constructed a system for short-term speculation in the foreign exchange market.
It is interesting to see how RL performs in forex markets compared to stock markets, because of it´s higher liquidity which makes it able for the agent to rapidly execute trades. So it´s exciting to analyze if the agent is able to gain additional return from that. 


#### 1.3.1 Model and Methodology

In order to make the agent profitable the authors use financial profit directly as a reward and base the system decision making on information with potential for price prediction.
To fulfill this aim Carapuço et al. use market data in the form of ticks $T_i$, which is put out whenever the market price updates prices. A tick $T_i$ can be explained as follows:

$$
T_i= [B_i  A_i   Vb_i    Va_i  ],
$$

with $B_i,A_i$ being the bid and ask price at the time $T_i$ was updated and $Vb_i,Va_i$ represent the volume of units traded at the prices of $B_i$  and $A_i$.
These ticks build a data set which represent the environment, while the agents frequency of decisions is controlled by a variable called time_skip. For example a time_skip of 5000 leads to an average decision of every two hours. This underpins the authors aim to construct a system for short term speculation.

##### 1.3.1.1 State signal

The state signal consist mainly of features that are extracted from the market data, the tick data, and are simple descriptive statistics: mean, maximum, standard deviation and minimum values for the different tick components bid/ask prices and volumes. While the ask price is replaced with the spread $S_i≡A_i-B_i$, since bid and ask prices are highly correlated.

What´s interesting to see is how the authors solve the curse of non-stationarity within the price distribution. This is due a underlying bias value that yields to strongly correlated mean, maximum and minimum bid price values form a given window, which makes learning more difficult for a neural network:

Forex State Signla.PNG

Imagine splitting the windows into left and right for training and validation set, the range of values clearly distinguish. 

To solve this problem, Carapuço et al suggest to subtract from the features (mean, maximum, standard deviant and minimum values for the different components of the tick data) an estimate of the underlying bias, which is the bid value of the current tick $B_i$. The result are much more stable distributed features:


Forex State SignlaII.PNG

What´s also interesting to see is that the authors use a encoding method to train the agent that he or she  is able to detect higher volume and volatility at certain hours of a trading day, which is due to when major trading places are open. It is done by including two input variables:

$$
time^1_i = sin(2{\pi}*\frac{seconds_i}{86400}),
$$

$$
time^2_i = cos(2{\pi}*\frac{seconds_i}{86400}),
$$

where $seconds_i$ is the time when $T_i$ was published converted to seconds. This way the cyclical nature of hours is transferred to the neural network.

##### 1.3.1.2 Action and Reward signal

According to the current position of the asset the agents possible actions are hold, do nothing, open a long position or open a short position.

The reward is measured in a way to face the problem of unrealized earnings. Because the system is not punished for holding into positions with negative unrealized profits, it keeps a position open until its unrealized profit gets positive values regardless the time it takes. Therefore, a reward signal which is only based on the obtained profit has its behavioral flaws, since it is not sure that the asset regains its former value. 

The solution proposed by the authors includes a new additional reward component. The new component is a reward for variation of unrealized profits, which is the difference between the unrealized profit in the state $S_t$ and the unrealized profit in state $S_{(t+1)}$. Therefore, profit is obtained with less risk. 

##### 1.3.1.3 The Q-network

The fully connected neural network includes 148 input neurons with 20 ReLU neurons for each of the three hidden layers which follow the input layer. 
ReLU (rectified linear unit) is a linear function, that outputs the input directly if it´s positive otherwise the output will be 0. The authors choose this type of function because of it´s ability of training multi-layer neural networks. These linear activation functions represent the action values.


#### 1.3.2 The empirical output and it´s evaluation

Empirical results were drawn from the period 2010 until 2016, which is split into a training, validation and testing period on EUR/USD data.
The testing was done using a rolling window to solve the problem of the non-stationarity of the forex market, since a too big change from the training/validation period to the test period will cause a weaker performance. 

Analysis of the results show that validation profits are not a good predictor of profits within the test period. A Pearson correlation of -0.28 at a p-value significance of 0.2 underpins this independence.

But since standard deviation of paths within the validation and the test period have a significant correlation, stable candidates can be selected based on the validation process. The candidates differ e. g. from the step size of the rolling window, the $timeskip_i$ variable or the discount rate $γ$.

By analyzing the equity, which is the current size of the account added to the value of currently open positions relative to the initial account size, the growth is relatively stable with a maximum drawdown of -9.9. It also seems that the system has problems with a sidewise trend or no clear trend of the market dynamic. 


#### 1.3.3 Learnings and key take aways

Nevertheless some drawdowns, such that then no trend can be observed from the agent, the authors were able to build a system for short-term speculations in the forex market, which yielded an equity growth of approximately 60% within the time period.  The results show that reinforcement learning can also be applied to the Forex market but needs to be adjusted e. g. in the sense of robustness. However the agent was able to be profitable by only using limited types of financial data.



# Section II:  Description of the A2C as implemented in the examples for the project assignment

## Introduction to the A2C algorithm.

The Actor-to-Critic (A2C) algorithm is an optimization of traditional algorithms, as the A2C algorithm takes the current level of values into account for updating it´s parameters.

The algorithm belongs to the asynchronous methods for deep reinforcement learning and was first proposed by Mnih et. al in 2016 [1].

According to the A2C logic the value of an action (the Q value) does not only depend on how good the action is, but of how better it can be. Or in other words the value of the current action taken does not only depend on it´s overall value but more on the relative value, relative to the current level of possible rewards in the given state. 

To account for this fact the A2C algorithms additionally ads an additional term which "punishes" the observed return by subtracting a baseline, so that the agent is able to distinguish between good returns and returns which are below the baseline. This is possible due to the relation:

$$\mathbb{E}_{a_t \sim \pi_{\theta}}{\nabla_{\theta} \log \pi_{\theta}(a_t|s_t) b(s_t)} = 0.$$

Therefore only the information which yields to high returns will be used in the model. Another advantage is the reduction of the variance such that the output is more stable. 



## Derivation of the method´s logic

Without the A2C, e. g. general Q-learning, the update step of the parameters can be written as:


$$
\boldsymbol{\theta}_{\pi} \leftarrow \boldsymbol{\theta}_{\pi} + \eta G_t \frac{\nabla \pi_{\boldsymbol{\theta}_{\pi}}(A_t | S_t)}{\pi_{\boldsymbol{\theta}_{\pi}}(A_t | S_t)}
$$

While with the A2C Algorithm we introduce a baseline $b(s)$ which tells us how good the current action is compared to the overall state value:


$$
\boldsymbol{\theta}_{\pi} \leftarrow \boldsymbol{\theta}_{\pi} + \eta G_t -b(s)\frac{\nabla \pi_{\boldsymbol{\theta}_{\pi}}(A_t | S_t)}{\pi_{\boldsymbol{\theta}_{\pi}}(A_t | S_t)}
$$

With a simple replacment of  $G_t$ with $R_{t+1} +v(s')$ we can represent the method as Temporal Difference learning (TD).

But what is the purpose for using the baseline? One reason is that it tells us  how much better the action value can be. Let´s look at an simple example of how that works:


##Example 1

---

In the example let´s pretend we work as a analyst in a private equity fund which has the policy of only making investments which yield 2% more as the cost of capital of an equivalent investment. 
For the next quarter there are 3 possible firms with individual returns and risk classes:

```python
import random

#3 investments
investments = [1, 2, 3]

# yearly returns
returns = [0.09, 0.02, 0.1]

#the associated risk, represented by the cost of capital
CoC = [0.06, 0.03, 0.05]

#the value add tells us what additional return we get above the cost of capital

value_add = []

zip_obj = zip(returns, CoC)
for returns_i, CoC_i in zip_obj:
  value_add.append(returns_i - CoC_i)


#as we said our baseline is a minimum return of 2% above the cost of capital

baseline = 0.02

#Now let´s randomly decide for a project and let´s see how the A2C logic helps us to seperate good investements from ones we should mitigate:

class trader():
  def choose_action():
    return random.choice(investments)

choose_invest = trader.choose_action()
return_firm   = returns[choose_invest-1]
net_return    = return_firm - baseline

if net_return > 0:
   decision = "profitable enough"
else:
   decision = "not profitable enough"
   

print(f"The choosen investment is the firm {choose_invest}, which yields a return of {return_firm*100}%.")
print(f"Since we want to look at the net return we subtract the baseline of {baseline*100}% from the return and get a value add of {round(net_return*100)}%.")
print(f"Therefore the selected firm is {decision}.")




```



---
Now we have seen how the baseline helps us to distinguish between decisions which are both profitable and good or not as good as we want them to be.

In the reinforcement learning context of course, there is the question of what should we use as a baseline? Well, of course we could also say that we select the baseline ex ante, but given that we maybe not know how the environments dynamics interact and definitely not know how the future will be we should use a baseline which is capable of learning.

An obvious and common solution of the selection problem of the baseline is the on-policy value function $V^{\pi}(s_t)$. By selecting the value function we do not only have a baseline which is capable of adjusting it´s value to the learning effect of the agent but it also has the effect of variance reduction.
 
By subtracting the baseline from the value we get the "advantage function" which tells us the relative benefit of taking a certain action at a certain time during a certain time step. 

Let´s see in another example how this works within a trading context:


##Example 2

---



In the example let´s pretend we are at the last trading day within a certain time period and the rewards for the actions buy, hold or sell are 10, 4 and -2, with a complete random and identically distributed policy.

For the sake of simplicity let´s assume we only trade one stock.

```python
import random

#The environment only consists of one stock of which we already own one, then the possible actions and rewards are:
actions_num = [0, 1, 2]
actions_name = ["buy","hold","sell"]
rewards = [10, 4, -2]

#trade with a random policy
class trader():
  def choose_action():
    return random.choice(actions_num)

#save the action and the reward
a_num = trader.choose_action()
a_name = actions_name[a_num]
r = rewards[a_num]

#as a baseline we can simply use the Q value, which can be simply computet if we assume identicly distributed probabilities:
baseline = sum(rewards)/len(rewards)

print(f"The baseline is drawn from the q value which yields, by identically distributed probability, a value of {baseline}")
print(f"that means if the trader decides to {a_name} a certain stock, the information used is not {r} but {r} subtracted by {baseline}, which is {r - baseline}")
```



---


As we have seen we have a baseline which is capable to learn, in other words it is allowed to change. It is no problem if the probability distribution of the actions is identically or not or if the action is chosen over a neural network which we will see later. 

Now let´s look how the learning and the adjusting of the parameters is done once we subtracted the baseline.

For the moment let´s stick to parametrized functions which calculate our state values. Within the reinforcement learning we can express the update via:


$$
\boldsymbol{\theta} \leftarrow \boldsymbol{\theta} + ηG_t\frac{\nabla \pi_{\boldsymbol{\theta}_{\pi}}(A_t | S_t)}{\pi_{\boldsymbol{\theta}_{\pi}}(A_t | S_t)}
$$

The parameter ${\theta}$ is our weight, this is the parameter we want to update since this weight decides which action will be performed. Since we use policy gradients we do not need any policy, e. g. epsilon-greedy, since the change of the weights, the parameters, directly influence the probability for each action.

As we see the update is done by the policy gradient and the learning rate $η$.

By including the baseline $b(s)$ and switching to TD(0) learning we get:

$$
\boldsymbol{\theta} \leftarrow \boldsymbol{\theta} + η((R_t +  γv_{\pi}(s')) - b(s)) \frac{\nabla \pi_{\boldsymbol{\theta}_{\pi}}(A_t | S_t)}{\pi_{\boldsymbol{\theta}_{\pi}}(A_t | S_t)}
$$

But how does this help the gradient to update the weights or the parameters in the right dircetion? To answer this question let´s clarify how what the gradient tells us:

Basically the gradient, which is the first derivative of a function, e. g. a loss function, tells us how much the function or the loss changes if the parameter is increased. So if we see that this action was a bad idea because it has a small reward (small meaning because it is below the baseline) we do not want the agent to perform this specific action more frequent. Also if the action was a good idea we want that the agent learns to to this specific action more often. 

Since we update a policy function, which delivers directly probabilities the differentiation after the learning looks like this: 

$$
\frac{\nabla \pi_{\boldsymbol{\theta}_{\pi}}(A_t | S_t)}{\pi_{\boldsymbol{\theta}_{\pi}}(A_t | S_t)}
$$

This Expression can be simply solved by the chain rule.

Therefore we need values which are the output of our policy function, which will be approximated by the neural network. However the output of the function is denoted as $z_i$. Then $z_i$ is derived by the weight $w_{ij}$, while $i$ stands for each future variable $s_i$ or for each neuron in a neural network $h_i$ and $j$ stands for each action.

But what about updating the baseline? 
We already said that the advantage of using the state value as the baseline is that it has also the ability to learn, let´s see how this is done and how it is different from the policy gradient update.

One difference is of course that the update of the parameters which yield to the state value are not clipped to the range [0,1] since they do not represent probabilities. This makes the update via the gradient even easier:

$$
\boldsymbol{\theta} \leftarrow \boldsymbol{\theta} + η((R_t +  γv_{\pi}(s')) - \hat V_{\pi,\theta_v}(s)){\nabla \hat V_{\pi,\theta_v}(s)}
$$

So the gradient information relies on the fact how far away the reward plus the estimation of the state value, which is the approximation of the value of the current state according to the TD(0) method, is from the observation or how high the loss is, which we want of course to be minimized. 


If we know include the loss function into the policy gradient update and denote the actor parameters as ${\theta}_{\pi}$ and the critic paramaters as ${\theta}_V$ , we combine everything we have talked so far:

The critc´s parameters - the value function - are upgraded as follows:

$$
\boldsymbol{\theta}_V \leftarrow \boldsymbol{\theta}_V + η_V(R_{t+1} +  γV_{\pi,\theta_v}(S_{t+1}) - \hat V_{\pi,\theta_v}(S_t)){\nabla \hat V_{\pi,\theta_v}(S_t)}
$$

While the actor´s parameters follow this update step:


$$
\boldsymbol{\theta}_{\pi} \leftarrow \boldsymbol{\theta}_{\pi} + η_{\pi}(R_{t+1} +  γV_{\pi,\theta_v}(S_{t+1}) - \hat V_{\pi,\theta_v}(S_t)) \frac{\nabla \pi_{\boldsymbol{\theta}_{\pi}}(A_t | S_t)}{\pi_{\boldsymbol{\theta}_{\pi}}(A_t | S_t)}
$$

Now we directly see the interaction between the actor and the critic.

The parameter ${\theta_V}$ within the value function, the critic, is also used within the update step of the policy parameter, the actor. 


Next we can use the baseline, our TD estimate, in order to perform the A2C approach.

As we already said, we subtract the baseline, the state value estimate, from the reward $R_t$ to obtain the error:

$$
error = R_t - v_{\pi}(s) 
$$

Finally we can complete the update step by simply multiply the obtained gradients, from the policy and the value function, with the error. By doing this the gradient tolds the agent in which direction he or she has to put the weights, obtained from the neural network, in order to select the action which relates to the error with a higher or lower probability. If the error is positive, meaning that the target is higher as the baseline, the probability for the related action will rise and vice versa.

The last step is updating the parameters with the gradient information we just derived.

Therefore we have a neural network for the policy, the actor that chooses the action, and a neural network for a function which estimates the state values and represents the baseline. Does this makes sense? Of course, since we want a system that has two networks which interact with each other but both should be also updated. Therefore two neural networks instead of two approximated function makes sense!

As we already said the output of the policy function is a probability, which is the reason why the additional argument $activation = "softmax"$ is used.

With the softmax function we reach the intended property of the output:


*   For each probability ${\theta}_{\pi_i}$ shall be $>= 0$
*   Further the sum of the probabilities shall be $\sum_{i=1}{\theta}_{\pi_i}$

The softmax function fulfills these requirements:

$${\pi_{\theta}}(a_i|s) = \frac{exp^{x^t{\theta}_i}}{\sum_jexp^{x^t{\theta}_j}}$$

As the value function, or the neural network for the state value does not need  inputs to be probabilities, we do not need the softmax activation function here.

###Action selection and gradient update


---

In our trading context the agent has the possible actions of buying, holding or selling a stock which is defined in the environment as the metrics [0,1,2].

Then the agent chooses the action according to the weights which are first introduced within the neural network of the policy (the actor). 

As we already know we value every action by comparing it to the baseline and use the gradient to adjust the weights. This adjustment is done via the differentiation:

$$
\frac{\nabla \pi_{\boldsymbol{\theta}_{\pi}}(A_t | S_t)}{\pi_{\boldsymbol{\theta}_{\pi}}(A_t | S_t)}
$$

The policy gradient implemented in the code uses the log-probability due to scaling reasons and to solve the "underflow issue" which means that the computer only uses a limited digit floating point representation of fractions which leed many probabilites to be 0.

The gradients are derived by the function $GradientTape$ which is a convinient function in order to obtain the gradients. Let´s see how the function does the job for us.
For example let´s consider the function we want to compute the gradient at $x = 2$ for the function $y = x^2$:

Since we can not only do this with hard coded numbers or constants but rather with the probabilites and values we can use the function to obtain our gradients which are necessary for the update step. 

Within the $GradientTape$ the "one hot" method is used. It helps us to focus on the chosen action in each step. It simply turns categorical variables, such as our action list, e. g. $[0,1,2]$ into a binary vector.

For one stock and the possible actions buy, sell or hold the stock the action list can be presented as $[0,1,2]$ but with the one hot trick we can also transform it into a vector. Then e.g. the action 0 (hold) can be transformed into $[1,0,0]$. Basically the one hot approach applies to the "active" action, so the chosen action a value, e. g. $1$ and the value $0$ to the not active actions.

Combining the one hot method with the logarithmized probabilities within the gradient collection we have a nice way to overgo numerical problems, focus on chosen action and improve the agents performance and stability.

While we also use the one hot trick for the value gradient, we do not need to logarithmize any value and simply can update it with  ${\nabla \hat V_{\pi,\theta_v}(S_t)}$.

But how can we receive the information in order to adjust the weights with the gradients within a trading context?

Well, we simply execute one trade and get the target, which is the next reward. If we look into the $multiStockEnvironment$ we see that the reward $R$ at time $t$ is calculated by the change of the portfolio value which you may want to name $wealth$ at  time $t$:

$$
R_t = Wealth_t - Wealth_{t-1}
$$


Since the output of the value function is a estimation of the state value we have to add the value of the next state to the reward, we call this the $target$ at the time step $t$:

$$
target_t = R_t +  γV_{\pi,\theta_v}(S_{t+1})
$$

Now we finally can compute the $error_t$ and see how the actor and the critic interact within the code:

$$
error_t = target_t - \hat V_{\pi,\theta_v}(S_t))
$$

That´s it actually! We just compare the return achieved by the actor which is the reward of the trade, e. g. the change of $Wealth_t$ if we bought Apple, add the state value of the next state and compare it to the current state value.

The last two steps belongs to the part in which we want the agent to learn. In other words we want to let the agent know how he has to change the weights, which are in turn responsible which action is performed at time step $t$.

We give her or him the information by simple multiplying the each gradient $\frac{\nabla \pi_{\boldsymbol{\theta}_{\pi}}(A_t | S_t)}{\pi_{\boldsymbol{\theta}_{\pi}}(A_t | S_t)}$ and ${\nabla \hat V_{\pi,\theta_v}(S_t)}$ with the $error_t$ and finally update the weigts within both networks.

```python
#@title
#Global Intializations such as the multi Stock Environment

import requests
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pylab as plt
import itertools
import random
import seaborn as sns
from ta import add_all_ta_features

import tensorflow as tf
from collections import deque

class multiStockEnvironment():
    
    def __init__(self, data, tickers, state_variables, initial_investment, state_memory, use_scaling):

        # the data we want to use
        self.data = data
        # tickers for the stocks which are in the data
        self.tickers = tickers
        # the variables we want to use for state information
        self.state_variables = state_variables

        # a list of all possible actions
        self.action_list = list(map(list, itertools.product([0,1,2], repeat = len(self.tickers))))

        # money we give the agent in the beginning
        self.initial_investment = initial_investment
        # cash which is owned by the agent during an episode
        self.cash = None
        # the number of stocks an agent is holding of each stock
        self.n_stocks = None
        # the current time step
        self.current_step = None
        
        # in the beginning the scaler is not fitted so we set it to false
        self.scaler_fitted = False
        # if we want to get scaled observations from the _get_obs()-method
        self.use_scaling = use_scaling
        
        # if we have no state memory, we collect observations of states by random playing, see the initial_play method below
        if len(state_memory) == 0:
            self.initial_play()
        else:
            # if we already have a state memory, we can use it
            self.state_memory = state_memory
            self.scaler_fitted = True


    # a function to set everything to start when beginning to play an episode
    def reset(self):
        # set the current time step to zero
        self.current_step = 0
        # initialize a list for the number of stock information
        self.n_stocks = [0] * len(self.tickers)
        # in the beginning we hold as much cash as we got for an initial investment
        self.cash = self.initial_investment
        self.stock_memory = []
        
        return self._get_obs()
    
    # a function to get the current state variables, the information how much cash the agent currently holds and the number of stocks 
    def _get_obs(self):
        # concatenate the information of state, cash and number of stocks
        curr_obs = np.concatenate([self.data[self.state_variables].iloc[self.current_step, :].values, np.array([self.cash]), np.array(self.n_stocks)])

        # as long as we did not experience enough observations for scaling the state information, we remember the observations as examples
        if not(self.scaler_fitted):
            self.state_memory.append(curr_obs)
        
        # if the scaler is fitted and we want to use scaled states, we standardize them
        if self.scaler_fitted and self.use_scaling:
            curr_obs = (curr_obs - np.mean(self.state_memory, axis = 0)) / np.std(self.state_memory, axis = 0)
        
        return curr_obs

    # a function to return the value of the stocks held by the agant and its cash reserves
    def _get_value(self):
        return (self.data[['adjClose_' + t for t in self.tickers]].iloc[self.current_step, :].values.dot(self.n_stocks) + self.cash)
    
    def _get_info(self):
        return self.n_stocks.copy(), self.cash
    
    # a function to execute trades corresponding to the current action by the agent
    def _trade(self, action):
        action_vec = self.action_list[action]

        # determine which stocks should be sold and bought
        sell_index = []
        buy_index = []

        for i, a in enumerate(action_vec):
            if a == 0:
                sell_index.append(i)
            if a == 2:
                buy_index.append(i)

        # if there is anything to sell, sell it to the current price and add this to your cash holdings
        if sell_index:
            # update the amount of cash available
            self.cash += self.data[['adjClose_' + self.tickers[s] for s in sell_index]].iloc[self.current_step, :].values.dot([self.n_stocks[i] for i in sell_index])

            # update the number of stocks you are holding
            for s in sell_index:
                self.n_stocks[s] = 0

        # if there are stocks to buy, split the available cash in equal fractions of money and buy as many stocks as possible for this money
        if buy_index:
            # determine the money per stock
            money_per_stock = self.cash / len(buy_index)
            # determine how many stocks can be bought
            n_buy_stocks = np.floor(money_per_stock / self.data[['adjClose_' + self.tickers[b] for b in buy_index]].iloc[self.current_step, :].values)

            # update the number of stocks in your posssession
            for i, b in enumerate(buy_index):
                self.n_stocks[b] += n_buy_stocks[i]

            # update the cash which has been used to buy the stocks
            self.cash -= n_buy_stocks.dot(self.data[['adjClose_' + self.tickers[b] for b in buy_index]].iloc[self.current_step, :].values)
        
        self.stock_memory.append(self.n_stocks)

    # a function to exececute the trade record values development and return current information
    def step(self, action):
        # get the current value
        previous_value = self._get_value()
        # go one time step forward
        self.current_step += 1
        # execute the trade
        self._trade(action)
        # derivet the new current value
        current_value = self._get_value()
        
        # calculate the reward as the change in portfolio value plus cash reserves
        reward = (current_value - previous_value) 
        
        # check if we are at the end of the episode
        if self.current_step == (len(self.data) - 1):
            done = True
        else:
            done = False
            
        # this is a placeholder with additional information we may want to add in the future
        current_n_stocks, current_cash = self._get_info()
        info = {'stocks_in_possession': current_n_stocks, 'cash_position': current_cash}
        
        return self._get_obs(), reward, done, info
        
     
    # a function to play an episode with random actions to collect state variables for scaling
    def initial_play(self):
        # initialize a memory for exemplary state observations
        self.state_memory = []
        # start the episode
        s = self.reset()
        done = False
        
        # until the episode execute random actions and go one time step forward
        while not(done):
            a = random.choice(np.arange(0, len(self.action_list)))
            s_, r, done, info = self.step(a)
                
        print(f'Played one episode with random actions to collect state and reward observations!')
        
        self.state_memory = np.array(self.state_memory)
        self.scaler_fitted = True



```

```python
#@title
# a class definition for the Deep-Q-Agent
class DeepQAgent():
    def __init__(self, input_dim, output_dim, optimizer, gamma = 1):
        # input dimension is the number of state variables which are used for estimating q-values
        self.input_dim = input_dim
        # output dimension is the number of actions for which q-valus are estimated
        self.output_dim = output_dim
        # discounting factor
        self.gamma = gamma
        
        # a network for estimating q-values
        self.q_network = tf.keras.Sequential([
            tf.keras.layers.InputLayer(input_shape = (self.input_dim)),
            tf.keras.layers.Dense(20, activation = 'relu'),
            tf.keras.layers.Dense(10, activation = 'relu'),
            tf.keras.layers.Dense(self.output_dim)
        ])
        
        # a copy of the network above which is used for estimating action values of the next-step-action pair
        # this network is only updated after a certain number of steps
        self.target_network = tf.keras.Sequential([
            tf.keras.layers.InputLayer(input_shape = (self.input_dim)),
            tf.keras.layers.Dense(20, activation = 'relu'),
            tf.keras.layers.Dense(10, activation = 'relu'),
            tf.keras.layers.Dense(self.output_dim)
        ])
        
        # the tensorflow optimizer we want to use
        self.optimizer = optimizer
        
        if case_number == '1':
          self.epsilon = 0.30
          self.epsilon_decay = 0.999

        elif case_number == '1.1':
          self.epsilon = 0.5
          self.epsilon_decay = 0.999

        elif case_number == "1.2":
          self.epsilon  = 0.5
          self.epsilon_decay = 0.6
        
          
        # minimum value for epsilon
        self.epsilon_min = 0.01
        
        # a dynamic list for memorizing loss values - divergence between target and predictions
        self.loss_memory = deque(maxlen = len(data_train))
        
        # copy the parameters of the q-network to the target network in the beginning
        self.update_target_network()
        # set target step to zero, it will be increased by one after each training step and parameters are copied after one hundred target steps
        self.target_step = 0
        
    # a function to copy parameters of the q-network to the target network
    def update_target_network(self):
        self.target_network.set_weights(self.q_network.get_weights())
        
    # select an action by epsilon-greedy-strategy
    def action(self, state):
        if random.uniform(0, 1) <= self.epsilon:
            return random.choice(np.arange(0, self.output_dim))
        else:
            return np.argmax(self.q_network(state.reshape(1, -1)).numpy()[0])
    
    # parameter update by target and prediction divergence
    def train(self, state, action, reward, next_state, done):
        # derive the target value
        if done:
            target = reward
        else:
            target = reward + self.gamma * np.amax(self.target_network(next_state.reshape(1, -1)).numpy()[0])
            
        # determine the gradient of the q-network with respect to the divergence between the target and the predicted q-value
        with tf.GradientTape() as tape:
            tape.watch(self.q_network.trainable_variables)
            # make prediction with the q-network
            predict = self.q_network(state.reshape(1, -1))
            # generate a target vector with all values equal to the prediction
            full_target = predict.numpy()
            # replace the real target value for the action which has been chosen
            full_target[0, action] = target
            full_target = tf.convert_to_tensor(full_target)

            # determine the loss - squared difference between target and prediction
            loss = tf.square(full_target - predict)
    
        # memorize the current loss value
        self.loss_memory.append(loss.numpy()[0])
    
        # get the gradients and use them for a gradient update step
        self.optimizer.apply_gradients(zip(tape.gradient(loss, self.q_network.trainable_variables), self.q_network.trainable_variables))

        # reduce epsilon after each gradient update
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay
            
        # after 100 training steps renew parameters in the target network by current parameters of the q-network
        self.target_step += 1
        if self.target_step == 100:
            self.update_target_network()
            self.target_step = 0
```

#### A2C Agent and training instance function

```python
#@title
# a class for the actor-critic agent
class A2C():
    def __init__(self, input_dim, output_dim, policy_optimizer, value_optimizer, action_list):
        # input dimension = number of variables used for police and state value estimation
        self.input_dim = input_dim
        # output dimension for actions
        self.output_dim = output_dim
        # an optimizer used for the policy network
        self.policy_optimizer = policy_optimizer
        # an optimizer used for the state value network
        self.value_optimizer = value_optimizer
        
        # a neural network for the policy
        self.policy_network = tf.keras.Sequential([
            tf.keras.layers.InputLayer(input_shape = self.input_dim),
            tf.keras.layers.Dense(20, activation = 'relu'),
            tf.keras.layers.Dense(10, activation = 'relu'),
            tf.keras.layers.Dense(self.output_dim, activation = 'softmax')
        ])
        
        # a neural network for state value estimation
        self.value_network = tf.keras.Sequential([
            tf.keras.layers.InputLayer(input_shape = self.input_dim),
            tf.keras.layers.Dense(20, activation = 'relu'),
            tf.keras.layers.Dense(10, activation = 'relu'),
            tf.keras.layers.Dense(1)
        ])
        
        # the list with all possible actions
        self.action_list = action_list
        
    #  a method for choosing an action according to current action probabilities
    def action(self, state):
        action = random.choices(np.arange(len(env_train.action_list)), weights = self.policy_network(state.reshape(1, -1))[0])[0]
        return action
    
    # a function to derive necessary gradients and update parameters of the policy and the value network
    def train(self, state, action, reward, next_state, done):
        
        # first get the gradients of the policy function for the chosen action
        with tf.GradientTape() as policy_tape:
            # the one-hot "trick" is used to focus on the chosen action in the step
            one_hot = np.zeros(shape = (len(self.action_list), 1))
            one_hot[action, 0] = 1.
            one_hot = tf.convert_to_tensor(one_hot, dtype = tf.float32)

            # this is the probability for choosing the current action according to the policy network function
            prob_a = tf.math.log(tf.linalg.matmul(self.policy_network(state.reshape(1, -1)), one_hot))

        # gradient of the log-probability from above with respect to the parameters of the policy network
        policy_grads = policy_tape.gradient(prob_a, self.policy_network.trainable_variables)
        
        # get the state value
        with tf.GradientTape() as value_tape:
            value_s = self.value_network(state.reshape(1, -1))

        # gradient of the state value with respect to the parameters of the state value network
        value_grads = value_tape.gradient(value_s, self.value_network.trainable_variables)
        
        # determine the return with TD(0) method
        if done:
            target = reward
        else:
            target = reward + self.value_network(next_state.reshape(1, -1)).numpy()[0][0]
                
        # how far away is the prediction of the state value from the (semi-)realized TD(0) return
        error = target - value_s.numpy()[0][0]
        
        # multiply each gradient value with the TD(0)-error
        policy_grads = [pg * error for pg in policy_grads] 
        value_grads = [vg * error for vg in value_grads] 
        
        # update all parameters according to the gradient information
        self.value_optimizer.apply_gradients(zip(value_grads, self.value_network.trainable_variables))
        self.policy_optimizer.apply_gradients(zip(policy_grads, self.policy_network.trainable_variables))
```

### Special Case I

Next to the variable adjustments and agent specific adjustments there is another opportunity to adjust the actions of the agent: The neural network (NN).

In order to see the sensitivities of the agent against his or her NN the following parameters are adjusted:

*   Number of layers
*   Number of neurons in the layers
*   Type of agent (A2C vs. DeepQ)


The variable setting, which was applied to both agent, consists of general measures such as the dcf and stock prices. Trend and volatility variables were not included so that it is possible to analyze the pure influence of the adjustment within the neural network.

The different types of network adjustments can be seen in Table 1 case number 3 to 3.3

To derive a meaningful answer the results of different adjustments were obtained from a 10 times for-loop which saved the returns from the test run that followed after the last training run. Then the 10 results were averaged. 

The adjusted A2C and DeepQ is implemented in the following and a exemplary run for the A2C algorithm with the adjustments of 3.1 (Table 1) is provided there after including the companies Linde, Oracle, Philip Morris and Delta Air Lines.

Further an examplary case for the training is provided below.  

```python
#@title
# a class for the actor-critic agent
class A2C_adjusted():
    def __init__(self, input_dim, output_dim, policy_optimizer, value_optimizer, action_list):
        # input dimension = number of variables used for police and state value estimation
        self.input_dim = input_dim
        # output dimension for actions
        self.output_dim = output_dim
        # an optimizer used for the policy network
        self.policy_optimizer = policy_optimizer
        # an optimizer used for the state value network
        self.value_optimizer = value_optimizer
        
        # a neural network for the policy
        self.policy_network = tf.keras.Sequential([
            tf.keras.layers.InputLayer(input_shape = self.input_dim),
            tf.keras.layers.Dense(25, activation = 'relu'),
            tf.keras.layers.Dense(15, activation = 'relu'),
            tf.keras.layers.Dense(15, activation = 'relu'),
            tf.keras.layers.Dense(self.output_dim, activation = 'softmax')
        ])
        
        # a neural network for state value estimation
        self.value_network = tf.keras.Sequential([
            tf.keras.layers.InputLayer(input_shape = self.input_dim),
            tf.keras.layers.Dense(25, activation = 'relu'),
            tf.keras.layers.Dense(15, activation = 'relu'),
            tf.keras.layers.Dense(15, activation = 'relu'),
            tf.keras.layers.Dense(1)
        ])
        
        # the list with all possible actions
        self.action_list = action_list
        
    #  a method for choosing an action according to current action probabilities
    def action(self, state):
        action = random.choices(np.arange(len(env_train.action_list)), weights = self.policy_network(state.reshape(1, -1))[0])[0]
        return action
    
    # a function to derive necessary gradients and update parameters of the policy and the value network
    def train(self, state, action, reward, next_state, done):
        
        # first get the gradients of the policy function for the chosen action
        with tf.GradientTape() as policy_tape:
            # the one-hot "trick" is used to focus on the chosen action in the step
            one_hot = np.zeros(shape = (len(self.action_list), 1))
            one_hot[action, 0] = 1.
            one_hot = tf.convert_to_tensor(one_hot, dtype = tf.float32)

            # this is the probability for choosing the current action according to the policy network function
            prob_a = tf.math.log(tf.linalg.matmul(self.policy_network(state.reshape(1, -1)), one_hot))

        # gradient of the log-probability from above with respect to the parameters of the policy network
        policy_grads = policy_tape.gradient(prob_a, self.policy_network.trainable_variables)
        
        # get the state value
        with tf.GradientTape() as value_tape:
            value_s = self.value_network(state.reshape(1, -1))

        # gradient of the state value with respect to the parameters of the state value network
        value_grads = value_tape.gradient(value_s, self.value_network.trainable_variables)
        
        # determine the return with TD(0) method
        if done:
            target = reward
        else:
            target = reward + self.value_network(next_state.reshape(1, -1)).numpy()[0][0]
                
        # how far away is the prediction of the state value from the (semi-)realized TD(0) return
        error = target - value_s.numpy()[0][0]
        
        # multiply each gradient value with the TD(0)-error
        policy_grads = [pg * error for pg in policy_grads] 
        value_grads = [vg * error for vg in value_grads] 
        
        # update all parameters according to the gradient information
        self.value_optimizer.apply_gradients(zip(value_grads, self.value_network.trainable_variables))
        self.policy_optimizer.apply_gradients(zip(policy_grads, self.policy_network.trainable_variables))


# a class definition for the Deep-Q-Agent
class DeepQAgent_adjusted():
    def __init__(self, input_dim, output_dim, optimizer, gamma = 1):
        # input dimension is the number of state variables which are used for estimating q-values
        self.input_dim = input_dim
        # output dimension is the number of actions for which q-valus are estimated
        self.output_dim = output_dim
        # discounting factor
        self.gamma = gamma
        
        # a network for estimating q-values
        self.q_network = tf.keras.Sequential([
            tf.keras.layers.InputLayer(input_shape = (self.input_dim)),
            tf.keras.layers.Dense(25, activation = 'relu'),
            tf.keras.layers.Dense(15, activation = 'relu'),
            tf.keras.layers.Dense(15, activation = 'relu'),
            tf.keras.layers.Dense(self.output_dim)
        ])
        
        # a copy of the network above which is used for estimating action values of the next-step-action pair
        # this network is only updated after a certain number of steps
        self.target_network = tf.keras.Sequential([
            tf.keras.layers.InputLayer(input_shape = (self.input_dim)),
            tf.keras.layers.Dense(25, activation = 'relu'),
            tf.keras.layers.Dense(15, activation = 'relu'),
            tf.keras.layers.Dense(15, activation = 'relu'),
            tf.keras.layers.Dense(self.output_dim)
        ])
        
        # the tensorflow optimizer we want to use
        self.optimizer = optimizer
        
        if case_number == '1':
          self.epsilon = 0.30
          self.epsilon_decay = 0.999

        elif case_number == '1.1':
          self.epsilon = 0.5
          self.epsilon_decay = 0.999

        elif case_number == "1.2":
          self.epsilon  = 0.5
          self.epsilon_decay = 0.6
        
          
        # minimum value for epsilon
        self.epsilon_min = 0.01
        
        # a dynamic list for memorizing loss values - divergence between target and predictions
        self.loss_memory = deque(maxlen = len(data_train))
        
        # copy the parameters of the q-network to the target network in the beginning
        self.update_target_network()
        # set target step to zero, it will be increased by one after each training step and parameters are copied after one hundred target steps
        self.target_step = 0
        
    # a function to copy parameters of the q-network to the target network
    def update_target_network(self):
        self.target_network.set_weights(self.q_network.get_weights())
        
    # select an action by epsilon-greedy-strategy
    def action(self, state):
        if random.uniform(0, 1) <= self.epsilon:
            return random.choice(np.arange(0, self.output_dim))
        else:
            return np.argmax(self.q_network(state.reshape(1, -1)).numpy()[0])
    
    # parameter update by target and prediction divergence
    def train(self, state, action, reward, next_state, done):
        # derive the target value
        if done:
            target = reward
        else:
            target = reward + self.gamma * np.amax(self.target_network(next_state.reshape(1, -1)).numpy()[0])
            
        # determine the gradient of the q-network with respect to the divergence between the target and the predicted q-value
        with tf.GradientTape() as tape:
            tape.watch(self.q_network.trainable_variables)
            # make prediction with the q-network
            predict = self.q_network(state.reshape(1, -1))
            # generate a target vector with all values equal to the prediction
            full_target = predict.numpy()
            # replace the real target value for the action which has been chosen
            full_target[0, action] = target
            full_target = tf.convert_to_tensor(full_target)

            # determine the loss - squared difference between target and prediction
            loss = tf.square(full_target - predict)
    
        # memorize the current loss value
        self.loss_memory.append(loss.numpy()[0])
    
        # get the gradients and use them for a gradient update step
        self.optimizer.apply_gradients(zip(tape.gradient(loss, self.q_network.trainable_variables), self.q_network.trainable_variables))

        # reduce epsilon after each gradient update
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay
            
        # after 100 training steps renew parameters in the target network by current parameters of the q-network
        self.target_step += 1
        if self.target_step == 100:
            self.update_target_network()
            self.target_step = 0
```

### Conclusion of Section III

The above analysed Base and Special Cases show that a well thought-out algorithm can contribute to the increase in value of a Portfolio. However the implemented algorithms are quiet volatilie regarding the ouptut. Additional layers in the neural network could partially solve the problem as seen in the last Base case or the second Special case. 

Nevertheless the right variable setting, neural network adjustment and epsilon/epsilon decay values have to be set by the user. Therefore the sole algorithm does not always yield the best results. 
Even though every issue which emerged during the empirical investigation, in the sense of a weak performance, could be partially solved by increasing variables and improving the neural network. However, this leads to overfitting and shows the sensitiviy of the algorithm against it´s input.
The literature and community suggest data driven decision making in order to set the right variables and values for the parameters, such as CV. 

However summarizing the positive results, it can be said that reinoforcement learning is a benefit for stock trading and portfolio optimization. Further the Deep Q algorithm may should only be used for short term speculations while the A2C agent is probably the more practical method, since it builds a broughter portfolio without the curse finding the right values, e. g.  for epsilon (decay), and still is able to contribute to the overall portfolio wealth development. Together with an appropriate neural network and trend capturing variables the A2C algorithm performed best in all cases.  

However, even if the cases show promising results, price prediction is still not 100% possible since they can be seen as random. The problem is simply that the algorithm gets trained with historical data and is then expected to make decisions for the future based on that. 

Of course in order to solve that several adjustments and variables can be included but the risk of falling into the Rube Goldberg syndrome then also increases, as some critics say [4]. So a very good algorithm should probably also include the underlying business, its capital structure, macro-economic fundamentals and also human emotions and intentions. The price is only the derivative of these elements. [4]

Nevertheless reinforcement learning in trading contexts is a sufficient gain for the financial market industry and can be used to solve several strategic problems within it.
